### Q) 치킨집이 가장 적은 지역은??? 

- 인구대비 치킨집 수가 적다

- Q3) 서울에서는?

```python
import pandas as pd
```

```python
# 앞에서 정리한 치킨집 매장수, 지역별 인구 데이터를 불러옵니다. 
chicken = pd.read_excel('./data/chicken.xlsx')
population = pd.read_excel('./data/population.xlsx', thousands=',')   # 천 단위 구분기호 , 
```

```python
chicken.info()
```

```python
chicken.head()
```

```python
# 광역시도, 시군구 통합 컬럼 만들기
chicken['시군구_병합'] = chicken['광역시도'] + ' ' + chicken['시군구']
```

```python
chicken.head(1)
```

```python
population.head(1)
```

```python
# 병합하기
total = pd.merge(left = chicken, 
                right = population, 
                how = 'left', 
                left_on = '시군구_병합', 
                right_on = '시군구정리')
```

```python
total.head()
```

```python
total.columns
```

- 인구 만명당 매장수 계산하기

```python
# 인구만명당 매장수 정리하기
for c in ['1년미만', '1~2년', '2~3년', '3~5년','5년 이상', '계']:
    name = c + '_만명당'
    total[name] = total[c] / (total['총인구수'] / 10000 )      # 인구수 만명당 치킨집 수
```

```python
total.columns
```

```python
# 필요한 컬럼만 선택하기
df = total[['시군구정리','1년미만_만명당', '1~2년_만명당', '2~3년_만명당','3~5년_만명당', '5년 이상_만명당', '계_만명당']]
df.head()
```

```python
# 정렬하기
df = df.sort_values(by = '계_만명당', ascending= False)
df.head(10)
```

```python
df.tail(10)
```

```python
# 그래프 그리기
import matplotlib
from matplotlib import font_manager, rc
import platform
import matplotlib.pyplot as plt
import seaborn as sns


# 이미지 한글 표시 설정
if platform.system() == 'Windows':  # 윈도우인 경우 맑은고딕
    font_path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=font_path ).get_name()
    rc('font', family=font_name)
else:    # Mac 인 경우 애플고딕
    rc('font', family='AppleGothic')

#그래프에서 마이너스 기호가 표시되도록 하는 설정입니다.
matplotlib.rcParams['axes.unicode_minus'] = False   
```

```python
### 상위 지역  비교하기!   by 히트맵! 
# 히트맵은 값 부분에 숫자만 !!!!   -->  시각화에 필요없는 컬럼은 제외 or 인덱스로 설정하기
```

```python
df.columns
```

```python
# '시군구정리'  컬럼 ->  인덱스로 
df_graph = df.set_index('시군구정리')  
df_graph
```

```python
# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(10,60) )
# 그래프 제목 붙이기 
plt.title('주민인구만명당치킨집수비교', fontsize=10)
# 히트맵 그리기
sns.heatmap(data = df_graph,  
            annot=True, fmt = '.1f', 
            cmap = "RdYlBu_r")

fig.savefig('./img/주민인구만명당치킨집수.png')
```

- 상위 지역만 살펴보기
- 지역별 치킨집수 합계는 제외하고 살펴보기

```python
# 계는 제외하고 연차만 비교하기...
df_graph_only_years = df_graph[df_graph.columns[:-1]]   # 마지막 컬럼 제외
df_graph_only_years
```

```python
# 주민인구대비 치킨집이 많이 밀집된 상위 20개 지역


# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(10,10) )
# 그래프 제목 붙이기 
plt.title('주민인구만명당치킨집수비교(Top20)', fontsize=10)
# 히트맵 그리기
sns.heatmap(data = df_graph_only_years.head(20),  
            annot=True, fmt = '.1f', 
            cmap = "RdYlBu_r")
fig.savefig('./img/만명당치킨집수비교(Top20).png')  # 저장하기
```

```python
# 주민인구대비 치킨집 가장 적은 지역 10개

# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(10,7) )
# 그래프 제목 붙이기 
plt.title('주민인구만명당치킨집수비교(Bottom10)', fontsize=10)
# 히트맵 그리기
sns.heatmap(data = df_graph_only_years.tail(10),  
            annot=True, fmt = '.1f', 
            cmap = "RdYlBu_r")
fig.savefig('./img/만명당치킨집수비교(Bottom10).png')  # 저장하기
```

### Add)  서울시만 ??

```python
df_graph_only_years.head()
```

```python
# 서울특별시 가  인덱스에 포함된 데이터 선택하기
cond = [ ]
select = '서울특별시'
for region in df_graph_only_years.index:
    if select in region:
        cond.append(True)
    else:
        cond.append(False)

data_seoul = df_graph_only_years[cond]
data_seoul.head()
```

```python
# 서울시 구별 주민인구수당 매장수 비교

# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(10,7) )
# 그래프 제목 붙이기 
plt.title('서울시 주민인구 만명당 치킨집수 비교', fontsize=10)
# 히트맵 그리기
sns.heatmap(data = data_seoul,  
            annot=True, fmt = '.1f', cmap = "RdYlBu_r")
fig.savefig('./img/서울시주민인구만명당치킨집수비교.png')  # 저장하기
```
