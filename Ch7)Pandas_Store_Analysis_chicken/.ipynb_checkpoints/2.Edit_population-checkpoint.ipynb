## 인구 데이터 불러오기

```python
import pandas as pd
```

```python
fname = './population/주민등록인구및세대현황_월간.xlsx'
```

```python
raw = pd.read_excel(fname, header = 2, thousands = ',')
raw.head()
```

인구 데이터 <- ->  카페 매장수 데이터 병합! 

- 인구데이터 지역 : `서울특별시 종로구 (1111000000)`


- 매장 데이터 지역 표기:  `서울특별시 / 종로구`


==> 지역 명칭 통일 필요

```python
# 행정기관 코드 없이  서울특별시 xx구 형식으로 정리하기
temp_list = [ ]
for c in raw['행정기관']:
    temp = c.split(' (')[0]
    temp_list.append(temp)
raw['시군구정리'] = temp_list    
```

```python
# 저장하기
raw.to_excel('./data/population.xlsx', index = False)
```
