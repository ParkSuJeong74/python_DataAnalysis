```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import MiniMap
import pandas as pd
```

folium.Marker([위도, 경도], 
             tooltip = 'xxx',
             icon = `folium.Icon(color = '색상')`)

- 색상명: 


    ‘red’, ‘blue’, ‘green’, ‘purple’, 
    ‘orange’, ‘darkred’, ’lightred’, 
    ‘beige’, ‘darkblue’, ‘darkgreen’, 
    ‘cadetblue’, ‘darkpurple’, ‘white’, 
    ‘pink’, ‘lightblue’, ‘lightgreen’, 
    ‘gray’, ‘black’, ‘lightgray’

```python
# 지도 생성하기
seoul_station = [37.5536067,126.9674308]  
m = folium.Map(location=seoul_station,
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)  # 클러스터 추가하기

# 미니맵 추가하기
minimap = MiniMap() 
minimap.add_to(m)

# 선별진료소 추가하기
raw = pd.read_excel('./data/선별진료소.xlsx')
for i in raw.index:
    name = raw.loc[i, 'place_name']
    long = raw.loc[i, 'x']
    lat = raw.loc[i, 'y']
    folium.Marker([lat, long],
                  tooltip= name,
                  icon=folium.Icon(color = 'red')
                 ).add_to(marker_cluster)    
    

    
# 선별진료소(승차검진) 추가하기
raw = pd.read_excel('./data/선별진료소(승차검진).xlsx')
for i in raw.index:
    name = raw.loc[i, 'place_name']
    long = raw.loc[i, 'x']
    lat = raw.loc[i, 'y']
    folium.Marker([lat, long],
                  tooltip= name,
                  icon=folium.Icon(color = 'blue')
                 ).add_to(marker_cluster)    
    

# 선별진료소(승차검진) 추가하기
raw = pd.read_excel('./data/임시선별검사소.xlsx')
for i in raw.index:
    name = raw.loc[i, 'place_name']
    long = raw.loc[i, 'x']
    lat = raw.loc[i, 'y']
    folium.Marker([lat, long],
                  tooltip= name,
                  icon=folium.Icon(color = 'green')
                 ).add_to(marker_cluster)        
    

m
```
