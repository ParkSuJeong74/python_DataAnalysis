```python
import folium
```

## 지도시각화

- 지도생성하기
    - `m` = folium.Map(location = [위도, 경도],  zoom_start = 확대정도)
        - 위도/y/Latitude
        - 경도/x/Longitude
        - zoom_start 확대정도: 0~ 18 (default 10)

#### ※ 참고) 배경지도(tile map) 선택하기
folium.Map(tiles = `"OpenStreetMap"`)
- "OpenStreetMap" (default)
- "Stamen Terrain"   /  "Stamen Toner"  /  "Stamen Watercolor"
- "CartoDB positron" / "CartoDB dark_matter"

```python
# 지도 생성하기
# 먼저, 지도를 펼쳐야 합니다.  지도에서 어느 부분을 살펴볼 것인지, 지도 중심의 좌표를 위도, 경도를 이용해 지정해야 합니다. 
# 서울역을 기준으로 살펴보겠습니다. 
# zoom_start 옵션을 통해, 처음에 얼마나 확대해서 볼 것인지를 지정할 수 있습니다. 
# 지도가 생성된 뒤에 위치 이동이나 확대/축소를 조정할 수도 있습니다. 

seoul_station = [37.5536067,126.9674308]
m = folium.Map(location=seoul_station,   # 기준좌표: 서울역)
               zoom_start=12)
m
```

```python
import pandas as pd
```

```python
raw = pd.read_excel('./data/선별진료소.xlsx')
raw.head(5)
```

```python
raw.info()
```

##### 정보 추가하기
- 마커 추가하기
        - folium.Marker([위도, 경도]).add_to(`m`)
- 원 추가하기
        - folium.CircleMarker([위도, 경도],radius= 원크기).add_to(`m`)
- 추가옵션: 
        - tooltip="마우스 올리면 보여질 정보"
        - popup="클릭하면 보여질 정보"          


```python
for i in raw.index:
    name = raw.loc[i, 'place_name']
    long = raw.loc[i, 'x']
    lat = raw.loc[i, 'y']
    folium.Marker([lat, long],   # [위도, 경도]
                  tooltip= name,  # 마우스 오버 표시 이름
              
                 ).add_to(m)    
m
```

```python
m.save('covid_test.html')
```

## ※ Cluster Marker 이용

```python
# MarkerCluster   라이브러리를 불러오겠습니다. 
from folium.plugins import MarkerCluster
from folium.plugins import MiniMap
```

```python
# 지도 생성하기
seoul_station = [37.5536067,126.9674308]  
m = folium.Map(location=seoul_station,
               zoom_start=12)
marker_cluster = MarkerCluster().add_to(m)  # 클러스터 추가하기

# 미니맵 추가하기
minimap = MiniMap() 
minimap.add_to(m)

for i in raw.index:
    name = raw.loc[i, 'place_name']
    long = raw.loc[i, 'x']
    lat = raw.loc[i, 'y']
    folium.Marker([lat, long],tooltip= name).add_to(marker_cluster)    
```

```python
m.save('./map/covid_test_cluster.html')
```
