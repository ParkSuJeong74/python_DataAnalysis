### 연도별 이름 보기

```python
raw.columns
```

```python
raw["YearOfBirth"].unique()
```

```python
raw.pivot_table(index="Name", columns="YearOfBirth",
               values="Number", aggfunc="sum")
```

```python
# 2004 => 2000년대
year=1915
year_group=str(year)[:-1] + "0년대"
year_group
```

```python
year_group_list = []
for year in raw["YearOfBirth"]:
    year_group=str(year)[:-1] + "0년대"
    year_group_list.append(year_group)
len(year_group_list)
```

```python
raw["Year_Group"] = year_group_list
raw.info()
```

```python
df=raw.pivot_table(index="Name", columns="Year_Group",
               values="Number", aggfunc="sum")
df.head()
```

```python
name="Catherine"
df[df.index==name]
```

```python
name="Jenny"
df[df.index==name]
```

```python
name_list = ['Mary','Luna','Kevin','Catherine', 'Mike','Simon', 'Chris','Jenny','Anthony', 'Felix']
cond = df.index.isin(name_list)
mj_name=df[cond]
mj_name
```

```python
mj_name.sum() # 세로로 더함
mj_name_sum=mj_name.sum(axis="columns")
mj_name_sum
```

```python
for column in mj_name.columns:
    mj_name[column]=mj_name[column] / mj_name_sum * 100
```

## 복습 : 연도별로 트렌드 보기

```python
import pandas as pd
```

```python
raw=pd.read_csv("./data/babyNamesUS.csv")
raw.head() # 확인
```

```python
year_group_list = []
for year in raw["YearOfBirth"]:
    year_group=str(year)[:-1] + "0년대"
    year_group_list.append(year_group)
len(year_group_list)
```

```python
year_group_list=[]
for year in raw["YearOfBirth"]:
    year_group=year//10 # int(year/10) 와 같음
    year_group=f"{year_group}0년대"
    year_group_list.append(year_group)
```

```python
raw["Year_Group"] = year_group_list
raw.head()
```

```python
df=raw.pivot_table(index="Name", values="Number",
                    aggfunc="sum", columns="Year_Group")
df.head()
```

```python
df.to_excel("Year_Name_Analysis.xlsx")
```

### 고급

```python
df["합계"]=df.sum(axis="columns")
df.head()
```

```python
year_group = "1940년대"
df.columns[:11]
```

```python
year_gropp_list=[]
for year_group in df.columns[:11]:
    df[f"{year_group}_비율"]=df[year_group] / df["합계"] *100
df.head()
```
