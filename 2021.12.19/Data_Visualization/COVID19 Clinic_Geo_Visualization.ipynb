```python
import folium
```

### 지도 만들기

```python
seoul_station = [37.5536387,126.9671714]
map=folium.Map(location=seoul_station, zoom_start=12) # 얼마나 가까이서 볼건지, 작을수록 멈
map
```

```python
# 마커하기 (마커 생성 후 지도에 추가)
folium.Marker(seoul_station,
             tooltip="Seoul",
             popup="Station").add_to(map)
map
```

### 선별 진료소 지도 만들기

```python
# 지도 만들기
seoul_station = [37.5536387,126.9671714]
map=folium.Map(location=seoul_station, zoom_start=12)

for i in raw.index:
    lat=raw.loc[i, 'y'] # 위도
    long=raw.loc[i, "x"] # 경도
    name=raw.loc[i, "place_name"]
    call=raw.loc[i, "대표 전화번호"]
    
    location= [lat,long]
    
    folium.Marker(location,
             tooltip=name,
             popup=call).add_to(map)
map
```

```python
# 지도 저장하기
map.save("./COVID_Clinic_GEO.html")
```

### 클러스터 추가

```python
from folium.plugins import MarkerCluster
from folium.plugins import MiniMap
```

```python
# 지도 만들기
seoul_station = [37.5536387,126.9671714]
map=folium.Map(location=seoul_station, zoom_start=12)

minimap=MiniMap()
minimap.add_to(map)

marker_cluster=MarkerCluster().add_to(map)

for i in raw.index:
    lat=raw.loc[i, 'y'] # 위도
    long=raw.loc[i, "x"] # 경도
    name=raw.loc[i, "place_name"]
    call=raw.loc[i, "대표 전화번호"]
    
    location= [lat,long]
    
    folium.Marker(location,
             tooltip=name,
             popup=call).add_to(marker_cluster)
map
```

```python
map.save("./COVID_Clinic_GEO_cluster.html")
```
